```python
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/3cEYpjA9oz9GiPac4AsH4n"
```

```python
playlist_URI = playlist_link.split("/")[-1].split('?')[0] 
```

```python
data = sp.playlist_tracks(playlist_URI)
print(data)
```

```python
data['items']
```

```python
# Extracting Album info 
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_date = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    album_element = {'album_id': album_id, 'album_name': album_name, 'album_release_date': album_release_date, 'album_total_tracks': album_total_tracks, 'album_url': album_url}
    album_list.append(album_element)
```

```python
album_list
```

```python
data['items'][0]['track']['artists']
```

```python
# Extracting Artist info

artist_list = []
for row in data['items']:
    for key , value in row.items():
        if key == 'track':
            for artist in value['artists']:
                artist_name = artist['name']
                artist_id = artist['id']
                artist_href = artist['href']
                artist_dict = {'artist_name': artist_name,
                               'artist_id': artist_id, 'artist_href': artist_href}
                artist_list.append(artist_dict)
```

```python
artist_list
```

```python
# Extracting song info
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id': song_id , 'song_name': song_name, 'song_duration': song_duration, 'song_url': song_url, 'song_popularity': song_popularity, 'song_added': song_added, 'album_id': album_id, 'artist_id': artist_id}
    song_list.append(song_element)
```

```python
song_list
```

```python
album_df = pd.DataFrame.from_dict(album_list)
album_df.head()
```

```python
album_df = album_df.drop_duplicates(subset=['album_id'])
```
